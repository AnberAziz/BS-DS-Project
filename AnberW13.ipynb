```python
# ==========================================
# ⚖️ Week 13: Ethics & Explainable AI (XAI) - FIXED
# Student: Anber Aziz (Roll No. 2225165090)
# ==========================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Install SHAP if missing
try:
    import shap
except ImportError:
    !pip install shap
    import shap

print("--- Step 1: Rapid Data & Model Restoration ---")
# Reload to ensure clean state
url = "https://raw.githubusercontent.com/basilatawneh/Students-Academic-Performance-Dataset-xAPI-Edu-Data-/master/xAPI-Edu-Data.csv"
df = pd.read_csv(url)

# Clean & Encode
df.rename(columns={'NationalITy': 'Nationality', 'raisedhands': 'raised_hands', 'VisITedResources': 'visited_resources'}, inplace=True)
class_map = {'L': 0, 'M': 1, 'H': 2}
df['class_label_encoded'] = df['Class'].map(class_map)
features = ['raised_hands', 'visited_resources', 'AnnouncementsView', 'Discussion',
            'StudentAbsenceDays', 'ParentAnsweringSurvey', 'Relation', 'gender']
X = pd.get_dummies(df[features], drop_first=True)
y = df['class_label_encoded']
feature_names = X.columns.tolist()

# Scale & Split
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Re-build & Train Model
model = Sequential([
    Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=15, verbose=0)
print("Model Restored.")

print("\n--- Step 2: SHAP Analysis (Modern Method) ---")

# FIX: Define a wrapper to ensure pure Numpy output
# This prevents TensorFlow tensors from confusing SHAP
def predict_wrapper(data):
    return model.predict(data)

# Use 'shap.Explainer' (The modern unified interface)
# We use a subset of training data as the background
background = X_train[np.random.choice(X_train.shape[0], 100, replace=False)]

# Create the Explainer
# 'algorithm="permutation"' is robust for Keras models
explainer = shap.Explainer(predict_wrapper, background, algorithm="permutation")

# Calculate SHAP values for 50 test samples
print("Calculating SHAP values (This takes ~30 seconds)...")
shap_values = explainer(X_test[:50])

print("\n--- Step 3: Visualizations ---")

# The modern explainer output is an object, so we access slicing differently:
# shap_values.values shape is (50, 8, 3) -> (Samples, Features, Classes)

# Plot 1: Beeswarm Summary for Class 2 (High Performers)
print("Plot 1: Feature Importance Summary (High Performance Class)")
plt.figure()
# We slice [:, :, 2] to get all samples, all features, for Class 2
shap.plots.beeswarm(shap_values[:, :, 2], show=False)
plt.title("Drivers of 'High' Performance")
plt.show()

# Plot 2: Bar Plot
print("\nPlot 2: Feature Importance Ranking")
plt.figure()
shap.plots.bar(shap_values[:, :, 2], show=False)
plt.title("Top Features by Importance")
plt.show()

print("\n✅ SHAP Analysis Complete.")
```
